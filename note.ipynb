```python
max_unsigned_int = 0
has_negative_int = False

num_cols = len(in_array)
for y in range(num_cols):
    num_rows = len(in_array[y])
    for x in range(num_rows):
        n = in_array[y][x]
        
        if n < 0:
            _n = -n
            max_unsigned_int = max(max_unsigned_int, _n)
            has_negative_int = True
        else:
            max_unsigned_int = max(max_unsigned_int, n)
#         print(type(n), n, max_unsigned_int)

longest_hex_repr = f'{max_unsigned_int:x}'
print(type(longest_hex_repr), longest_hex_repr)

hex_repr_length = len(longest_hex_repr)

print(has_negative_int, hex_repr_length)

for y in range(num_cols):
    num_rows = len(in_array[y])
    print('[', end='')
    for x in range(num_rows):
        last_cell = (x == (num_rows - 1))
        n = in_array[y][x]
        if n < 0:
            _n = -n
            hex_repr = f'{_n:x}'
            hex_repr = hex_repr.zfill(hex_repr_length)
            hex_repr = f'-0x{hex_repr}'
        else:
            hex_repr = f'{n:x}'
            hex_repr = hex_repr.zfill(hex_repr_length)
            if has_negative_int:
                hex_repr = f'+0x{hex_repr}'
            else:
                hex_repr = f'0x{hex_repr}'
        if last_cell:
            print(hex_repr, end='')
        else:
            print(hex_repr, end=', ')
    print('],')
```
